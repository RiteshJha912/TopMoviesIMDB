```python
def gross(gross):
    if gross>=7.0:
        return "Excellent"
    elif gross>=6.0:
        return "Good"
    else:
        return "Average"
    
```

```python
data['gross_cat']=data['gross'].apply(gross)
```

```python
list1=[]
for value in data['genre']:
    list1.append(value.split(','))
```

```python
list1
```

```python
one_d=[]
for item in list1:
    for item1 in item:
        one_d.append(item1)
```

```python
one_d
```

```python
uni_list=[]
for item in one_d:
    if item not in uni_list:
        uni_list.append(item)
```
